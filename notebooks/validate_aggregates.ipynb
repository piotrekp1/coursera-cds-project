```python
import pandas as pd 
total_aggr = pd.read_hdf('../data/processed/train/aggr_train.hdf')
total = pd.read_hdf('../data/processed/train/total.hdf')
```

```python
total_aggr_6 = total_aggr[
    (total_aggr['year'] == 2015) & 
    (total_aggr['month'] == 6)
]
total_6 = total[
    (total['year'] == 2015) & 
    (total['month'].isin([3,4,5]))
]
```
