```python
import datetime as dt
import numpy as np
from dateutil.relativedelta import relativedelta

def get_null_df(index_names, index, columns):
    index = np.asarray(index).T.tolist()
    df = pd.DataFrame(index=index, columns=columns)
    df.index.set_names(index_names, inplace=True)
    return df

def get_null_date_df( date_from, date_to, columns, keys, key_names):
    dates = []
    while date_from != date_to:
        dates.append([date_from.year, date_from.month])
        date_from = date_from + relativedelta(months=1)
    
    indexes = [
        date + key 
        for date in dates
        for key in keys
    ]
    
    return get_null_df(['year', 'month'] + key_names, indexes, columns)
```

```python
df = get_null_date_df(dt.datetime(2014, 1, 1), dt.datetime(2014, 5,1), ['klops', 'mops'], [['a','bc'], ['druga_moz1', '2']], ['index1', 'index2'])
df
```
