```python
import pandas as pd 
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
df_train = pd.read_hdf('../data/processed/train/aggr_train_zeros.hdf')
df_train.head()
```

```python
df_test = pd.read_hdf('../data/processed/test/aggr_test.hdf')
df_test.head(10)
```

```python
df_train['date'] = pd.to_datetime(df_train['year'].astype(str) + '-' + df_train['month'].astype(str) + '-1')
df_train['date'].max()
```

```python
df_total = df_train 
df_total['train'] = 1
df_test['train'] = 0 
df_total = pd.concat([df_train, df_test], copy=False)
df_total.head()
```

```python
dim_red = pd.read_hdf('../data/processed/total_pca.hdf', 'pca')
```

```python
df_total[dim_red.columns] = dim_red
```

```python
df_train = df_total[df_total['train'] == 1]
df_test = df_total[df_total['train'] == 0]
```

```python
df_train[['ID'] + ['pca_' + str(i) for i in range(4)]].to_hdf('../data/processed/train/pca.hdf', 'pca')
df_test[['ID'] + ['pca_' + str(i) for i in range(4)]].to_hdf('../data/processed/test/pca.hdf', 'pca')
```

```python
df_total_cnt = df_total[[col for col in df_total.columns if col.startswith('count')]]
```

```python
df_total_cnt.to_hdf('../data/processed/total_cnt.hdf', 'total')
```
