```python
df_daily = df_train[
    ['year', 'month', 'day', 'weekday', 'item_cnt_day']
].groupby(['year','month', 'day']).agg([np.max, np.sum])

df_daily['wday'] = df_daily['weekday']['amax']
df_daily['item_cnt'] = df_daily['item_cnt_day']['sum']
df_daily = df_daily.drop(['weekday', 'item_cnt_day'], axis=1)
df_daily.columns = ['weekday', 'item_cnt_day']
```

```python

df_daily = df_daily.join(
    df_daily.groupby('weekday', as_index=False).mean(),
    on='weekday', rsuffix= '_mean',
)
```

```python
def days_from(cnt_date):
    # days from cnt_date event in this year (ignores cnt_date year)
    def f_days_from(date):
        return (date - cnt_date).days % 365
    return f_days_from
def get_date_norm_distr(center_date, wideness):
    # gets weight for the date based on the distance to the event in this year
    def get_value(date):
        days_dist = days_from(center_date)(date)
        return norm.pdf( days_dist / wideness) / norm.pdf(0)
    return get_value
```

```python
# new year
df_train['is_new_year'] = \
    ((df_train['month'] == 12) & (df_train['day'] > 26)) | \
    ((df_train['month'] == 1) & (df_train['day'] < 8))

new_years_eve = date(2014, 12, 31)
df_train['new_year_gaussian'] = df_train[df_train['month'].isin([1,12])]['date'].apply(get_date_norm_distr(new_years_eve, 5))

df_train['from_new_year_days'] = df_train['date'].apply(days_from(new_years_eve))
df_train['to_new_year_days'] = 365 - df_train['from_new_year_days']
```

```python
# feb_special
df_train['is_feb_special_time'] = \
    (df_train['month'] == 2) & (df_train['day'] > 20) & (df_train['day'] < 24)

special_feb_time = date(2014, 2, 22)

df_train['feb_special_gaussian'] = df_train[df_train['month'] == 2]['date'].apply(get_date_norm_distr(special_feb_time, 1))

df_train['from_feb_special_days'] = df_train['date'].apply(days_from(special_feb_time))
df_train['to_feb_special_days'] = 365 - df_train['from_feb_special_days']
```

```python
# nov_special
df_train['is_nov_special_time'] = \
    (df_train['month'] == 10) & (df_train['day'] > 2) & (df_train['day'] < 9)

special_nov_time = date(2014, 10, 6)

df_train['nov_special_gaussian'] = df_train[df_train['month'] == 10]['date'].apply(get_date_norm_distr(special_nov_time, 1))

df_train['from_nov_special_days'] = df_train['date'].apply(days_from(special_nov_time))
df_train['to_nov_special_days'] = 365 - df_train['from_nov_special_days']
```

```python
df_train.fillna(0, inplace=True)
```

```python
df_train['is_holiday'] = df_train['month'].isin([7, 8])
```
