```python
import pandas as pd 
df = pd.read_hdf('../data/processed/cumdata.hdf')
df.head()
```

```python
def join_max_cumcount(df, key):
    df_max_cumcount = df.groupby(key, as_index=False)['cumcount'].max()
    df_max_cumcount = df_max_cumcount[key + ['cumcount']]
    df = pd.merge(df, df_max_cumcount, on=key, suffixes=('', '_max'))
    return df
```

```python
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-1')
```

```python
df_max_min_date = df.groupby(key, as_index=False)['date'].min()
```
